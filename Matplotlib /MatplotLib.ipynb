## Task Histogram

```python
set(pdSecData.continent)
```

```python
YearIn1997 = pdSecData[pdSecData.year == 1997]
America1997 = YearIn1997[YearIn1997.continent == 'Americas']
Europe1997 = YearIn1997[YearIn1997.continent == 'Europe']
print(America1997.lifeExpectancy.mean())
print(Europe1997.lifeExpectancy.mean())
print(America1997.lifeExpectancy.median())
print(Europe1997.lifeExpectancy.median())
```

```python
# subplot(211)
# 1st is row, 2nd is column, 3rd is which one is next hist statement belongs
bins = 20
plt.subplot(2, 1, 1)
plt.title("Distribution of Life Expectancy")
plt.hist(America1997.lifeExpectancy, bins, range=(50, 80), edgecolor='black')
plt.ylabel('America')
plt.subplot(2, 1, 2) #subplot(212)
plt.hist(Europe1997.lifeExpectancy, bins, range=(50, 80), edgecolor='black')
plt.ylabel('Europe')
plt.show()
```

## Task - Line Graph

```python
set(pdSecData.country)
```

```python
us = pdSecData[pdSecData.country == 'United States']
china = pdSecData[pdSecData.country == 'China']
```

```python
usScalarPop = us.population / us.population.iloc[0] * 100
chinaScalarPop = china.population / china.population.iloc[0] * 100
```

```python
usScalarPop
```

```python
chinaScalarPop
```

#### Absolute Term

```python
plt.plot(us.year, us.population / 10**7)
plt.plot(china.year, china.population / 10**7)
plt.show()
```

#### Relative Term

```python
plt.plot(us.year, usScalarPop)
plt.plot(china.year, chinaScalarPop)
plt.show()
```
